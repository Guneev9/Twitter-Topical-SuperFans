```python
#Necessary libraries to be imported for UI to run
from ipywidgets import interact, interactive, fixed, interact_manual
from ipywidgets import *
import ipywidgets as widgets
from IPython.display import display
import SuperFans_Snippet as sfs
from ipywidgets import Button, Layout
import pandas as pd
from IPython.display import Markdown
```

```python
file_path = ''
home_path = ''
file = ''
selectBoxVal = ''

# Get all relavent file paths...
def getDirectoryPath(topic):
    global file_path, home_path, file
    
    file_path= '/' + (topic) + '_SuperFans_'
    
    if not os.path.exists((topic) + '_SuperFans'):
        os.makedirs((topic) + '_SuperFans')

    home_path=(topic) + '_SuperFans_'  + '/'
    
    file='parent_Tweets_' + (topic) + '.json'
    
    
# Method responsible for integrating the extraction methods that take care of extracting all the data from twitter api and remodeling them...
def handleTopicSubmit(t):
    if text.value:
        print("Extracting Data for topic ({})....".format(text.value))
        topic = text.value[1:]
        getDirectoryPath(topic)
        #tweet means hashtag
        print(".........................tweets extraction phase for topic {}............................".format(text.value))
        sfs.Tweets_Extraction(text.value)
        
        print(".....................................retweeters extraction phase................................")
        sfs.Twitter_SuperFans(file,text.value)
        
        print(".....................................Top 10 SuperFans of parent user's................................")
        sfs.Top10_influencers_of_parent(home_path,topicHashTag)

        print(".....................................Top 10 SuperFans................................")
        sfs.Top10_Influencers(home_path,topicHashTag)
        
        # Perform analysis...
        performExistingAnalysis(text.value)
    else :
        print('Text field cannot be blank')
    print('-----------------------------------------------------------')
    
    
# Method responsible for integrating initial method performing analysis for a given hastag when data has already been extracted
def performExistingAnalysis(topicHashTag):
    if topicHashTag:
        
        global selectBoxVal
        selectBoxVal = topicHashTag
        
        #fetching tag name from topicHashtag
        topic = str(topicHashTag[1:])
        
        #getting the directory path of the hashtag
        getDirectoryPath(topic)
        
        print("Performing analysis on existing data ({})....".format(topicHashTag))
        
        top_users =  sfs.twitter_visualizations(topicHashTag)
        
        #creation of drop down for exisitng tags
        createUserDropDown(top_users)
        
        #for displaying the word cloud
        sfs.twitter_cloud(topicHashTag)

    
# Method responsible for creating a drop down containing all the parent users as options
def createUserDropDown(top_users):
    
    printmd('**Select parent user from an existing list:**')
    
    selectBox = widgets.Select(
    options= (top_users),)

    interact(populateChildUserGraph, Users=top_users)

# Method responsible for creating a sub user graph corresponding to a single parent user selected from the dropdown
def populateChildUserGraph(Users):
    if Users:
        sfs.graph_based_on_parent_user(Users, selectBoxVal)
    
# Custom method used to design the text as bold or giving extra styling
def printmd(string):
    display(Markdown(string))

# Method used as a startup function which is responsible for loading all the initial UI components
def startUp():
    
    text = widgets.Text()
    printmd('**Select new HASHTAG TOPIC to extract data and perform analysis...**')
    display(text)

    button = widgets.Button(description="Search with new topic")
    display(button)

    printmd('**OR Select any TOPIC from dropdown to check analysis...**')

    #dictionary of exisitng topics to select from 
    Trending_Topics= {0:'', 1: '#WrestleMania', 2: '#TaxDay', 3: '#Metoo'}
    
    style = {'description_width': 'initial'}
    
    selectBox = widgets.Select(
        options=list(Trending_Topics.values()),
        value='#WrestleMania',
        description='Select topics from an existing list:', style=style,
    )

    interact(performExistingAnalysis, topicHashTag=list(Trending_Topics.values()))
    
    #returning values
    return text, button



        
```

```python
# Calling startup method to show all the User Interaction Components below
text, button = startUp()
button.on_click(handleTopicSubmit)
```
